```python
import numpy as np
import pandas as pd
import random

from matplotlib import pyplot as plt
from scipy.io import loadmat

plt.style.use("duarte.mplstyle")
%matplotlib widget
```

```python
xs_all = loadmat("../../../xs_all_noise.mat")
```

```python
xs_sanches1 = pd.DataFrame(xs_all["xs_all"][0][random.randrange(10,900)].T).drop(columns=[3,4,5,6,7,8,9,10,11]).rename(columns={0: "x", 1: "y", 2: "z"})
xs_sanches2 = pd.DataFrame(xs_all["xs_all"][0][random.randrange(10,900)].T).drop(columns=[3,4,5,6,7,8,9,10,11]).rename(columns={0: "x", 1: "y", 2: "z"})

display(xs_sanches1.tail(3))
display(xs_sanches2.tail(3))

cm = 1/2.54
fig, ax = plt.subplots(figsize=(16*cm, 12*cm),
                       subplot_kw=dict(projection='3d'),
                       gridspec_kw=dict(top=1, left=0, right=1, bottom=0),
                       constrained_layout=True)
ax.plot(xs_sanches1.x, xs_sanches1.y, xs_sanches1.z)
ax.invert_zaxis()
ax.set_zlabel("$z(t)$")
ax.set_xlabel("$x(t)$")
ax.set_ylabel("$y(t)$")
# ax.view_init(15,-30)

# fig.tight_layout()
fig.savefig("../../report/figures/4results/uav/trajectory.pdf",
            bbox_inches="tight")
```
