```python
import numpy as np
import pandas as pd
import random

from matplotlib import pyplot as plt
from scipy.io import loadmat

plt.style.use("../models/model_alpha/duarte.mplstyle")
%matplotlib widget
```

```python
XS_FILE_PATH = "../../../matlab_files/xs_all.mat"
xs_all = loadmat(XS_FILE_PATH)
```

```python
random.seed(1)
xs_sanches1 = pd.DataFrame(xs_all["xs_all"][random.randrange(10,900)][0].T).drop(columns=[3,4,5,6,7,8,9,10,11]).rename(columns={0: "x", 1: "y", 2: "z"})
xs_sanches2 = pd.DataFrame(xs_all["xs_all"][random.randrange(10,900)][0].T).drop(columns=[3,4,5,6,7,8,9,10,11]).rename(columns={0: "x", 1: "y", 2: "z"})

display(xs_sanches1.tail(3))
display(xs_sanches2.tail(3))

CM = 1/2.54
fig, ax = plt.subplots(figsize=(16*CM, 12*CM),
                       subplot_kw=dict(projection='3d'),
                       gridspec_kw=dict(top=1, left=0, right=1, bottom=0),
                       constrained_layout=True)
ax.plot(xs_sanches1.x[600:], xs_sanches1.y[600:], xs_sanches1.z[600:])
ax.invert_zaxis()
ax.set_zlabel("$z(t)$")
ax.set_xlabel("$x(t)$")
ax.set_ylabel("$y(t)$")
# ax.view_init(15,-30)

# fig.tight_layout()
fig.savefig("../../report/figures/4results/uav/trajectory.pdf",
            # bbox_inches="tight",
            backend="pgf")
```
