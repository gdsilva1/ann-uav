```python
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
plt.style.use("duarte")
%matplotlib widget
```

```python
def fi(n):
    if n == 1:
        return [1]
    if n == 2:
        return [1,1]
    if n > 2:
        list = [1,1]
        for _ in range(2, n):
            list.append(list[-1]+list[-2])
        return list
```

```python
cm = 1/2.54
SIZE = (12*cm, 6*cm)
```

```python
x = np.arange(-2,1,0.1)
```

```python

fig, ax = plt.subplots(figsize=SIZE)

ax.plot(x,x**2, zorder=1)
ax.scatter(x,x**2, marker=".", zorder=1),
ax.scatter(0, 0, zorder=2)
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.spines[["right", "top"]].set_visible(False)
# ax.annotate("Minimum",
#             xy=(14,2),
#             xytext=(10,200),
#             arrowprops=dict(facecolor="black", 
#                             arrowstyle="->",
#                             linewidth=0.5))
fig.tight_layout()
fig.savefig("../../../report/figures/2review/nn/gradient_descent.pdf", backend="pgf")
```

```python
xx = np.arange(0,15,1)
yy = 2*xx
```

```python
x_noise = [i + random.randrange(0,10) for i in xx]
y_noise = [i + random.randrange(0,10) for i in yy]
```

```python
figg, axx, = plt.subplots(figsize=SIZE)

axx = sns.regplot(x=x_noise, y=y_noise, ci=None, marker=".", line_kws={'linewidth':1}, scatter_kws={"color": "black"})
axx.set_xticks([])
axx.set_yticks([])
axx.spines[["right", "top"]].set_visible(False)
# plt.scatter(x_noise, y_noise)
figg.tight_layout()
figg.savefig("../../../report/figures/2review/nn/mae_chart.pdf", backend="pgf")
```
